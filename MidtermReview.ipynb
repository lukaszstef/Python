```python
def sumUnless(x, y):
    sum = x + y
    if 90<=sum<=100:
        print(100)
    else:
        print(sum)

sumUnless(10, 6)
sumUnless(10, 75)
sumUnless(80, 12)
```

## 2. Write a function that takes a list of words and counts the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings.

```python
def firstLast(words):
    count = 0
    for word in words:
        if (len(word) > 1) and (word[0] == word[-1]):
            count += 1
    print(count)
    
firstLast(['abba', 'wow', 'wwwoooowwww', '1221', 'no', 'maybe'])
```

## 6. Write a function that takes two lists and returns True if they have at least one common member.

```python
def inCommon(list1, list2):
    commons = []
    for i in list1:
        if i in list2:
            commons.append(i)
    if len(commons)>0:
        print("True")
    else:
        print("no common members")
```

```python
inCommon([1,2,3,4,5], [5, 6, 7])
inCommon([1,2,3,4,5], [6, 7])
inCommon(["welp", "nope", "yes"], ["if", "then", "yes"])
```

## 7. Write a function that counts the number of elements in a list within a specified range


```python
def withinRange(thisList, startNum, endNum):
    elemCount = 0
    for i in thisList:
        if i in range(startNum, endNum):
            elemCount += 1
    print(elemCount)

withinRange([3, 4, 5, 7, 10, 12, 30, 50, 66], 8, 20)
```

## 1. Create a function to get a string made of the first 2 and the last 2 chars from a given a string. If the string length is less than 2, return instead of the empty string.

```python
def stringOf2s(stringy):
    if len(stringy) < 2:
        print("  ")
    else:
        print(stringy[0:2] + stringy[-2:])

stringOf2s('inevitable')
stringOf2s('A')
stringOf2s('no')
```

## 2. Write a function to get a string from a given string where all occurrences of its first char have been changed to a dollar sign, except the first character itself.

```python
def change_char(givenString):
    firstLetter = givenString[0]
    modString = givenString.replace(firstLetter, '$')
    modString = firstLetter + modString[1:]
    print(modString)

change_char('inevitable')
```

## 3. Write a function to get a single string from two given strings, separated by a space and swap the first two characters of each string.

```python
def stringySwap(string1, string2):
  firstPart = string2[:2] + string1[2:]
  secondPart = string1[:2] + string2[2:]
  print(firstPart + ' ' + secondPart)

stringySwap('errors', 'language')
```

## 4. Write a function that adds 'ing' at the end of a given string (length should be at least 3). If the given string already ends with 'ing' then add 'ly' instead. If the string length of the given string is less than 3, leave it unchanged.

```python
def addEnds(givenString):
    if len(givenString) > 2:
        if givenString[-3:] == 'ing': 
            givenString += 'ly'
        else:
            givenString += 'ing'
    print(givenString)
    
addEnds('rock')
addEnds('sing')
addEnds('no')
```

## 5. Write a function that converts a given string to all uppercase if it contains at least 2 uppercase characters in the first 4 characters.

```python
def uppercaseIt(thisString):
    upperinfirst4 = 0
    for letter in thisString[:4]:
        if letter.isupper()==True:
            upperinfirst4+=1
    if upperinfirst4>=2:
        print(thisString.upper())
```
