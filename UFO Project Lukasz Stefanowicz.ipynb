```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#import libraries for qualitative,quantitative analysis and data manipulation.
```

```python
scrubbed=pd.read_csv('scrubbed.csv',low_memory=False)
#Read in the scrubbed csv.The scrubbed csv was a filtered complete csv from NUFORC.(An American UFO sighting agency )
```

```python
scrubbed
#read in the dataset by itself.
```

```python
NSDUH=pd.read_csv('NSDUHsaeExcelTab01-2016.csv',skiprows=5)
NSDUH
#The NSDUH DATASET is a csv file of illicit drug use in the past month from the year 2016.
#Skiprows=5 was used because the first five rows were the csv description.
#used to join State names.
```

```python
#2019 Census was used
Census=pd.read_csv('2019_Census.csv')
Census
#Census was read in and displayed.
```

```python
#Stateabb csv was read in.
stateabb=pd.read_csv('State-Abbrev.csv')
stateabb
```

```python
scrubbed['state']=scrubbed['state'].str.upper()
#to capitalize the state names
```

```python
scrubbedabb=scrubbed.merge(stateabb,left_on='state',right_on='Code')
scrubbedabb
#first merge(scrubbed UFO data to stateabb joined by state and code by their common key column)
```

```python
scrubbedabb=scrubbedabb.merge(Census,left_on='State',right_on='STATE')
scrubbedabb
#second merge.All the datasets had slightly different state columns and they were all merged on a 'State' column in their csv's.
```

```python
UFO=scrubbedabb.merge(NSDUH,left_on='State',right_on='State')
for i in UFO.columns:
    if UFO[i].dtype=='object':
        UFO[i]=UFO[i].str.upper()
#third merge NSDUH data was merged on 'State'.
#All the state columns were merged for the purpose to create a State column that was abbreviated with all the data from the
#multiple datasets.
```

```python
UFO
#UFO was created in the previous cell and was run here.
```

```python
UFO.dtypes
#dtypes was used to view the dataset's datatypes.
```

```python
UFO[['state','shape']].describe()
#describe was used on state and shape.The top state for sightings was CA.
#The top shape was light.Light was the top shape but doesnt really describe a shape so I used triangle for some of my findings.
```

```python
stategroups=UFO.loc[UFO['shape']=='TRIANGLE']
#A new table was created using loc to look up a State's Triangle sightings.
```

```python
stategroups=stategroups.groupby([UFO['state'],UFO['shape']])['shape'].count()
stategroups=stategroups.nlargest(10)
#states grouped by frequency and triangle shape
```

```python
stategroups.plot(kind='bar')
#bar graph of the top 10 states for triangle shaped UFOS.
```

```python
COshapes=UFO.loc[UFO['state']=='CO']
COshapes=COshapes.groupby([UFO['state'],UFO['shape']])['state'].count()
COshapes.plot(kind='bar')
#A table was created using loc the column 'state' and was ploted using a bar graph to show the top UFO shapes in Colorado.
```

```python
sightingcounts=UFO['state'].value_counts().to_frame()
sightingcounts=sightingcounts.reset_index()
sightingcounts.columns=['state','count']
sightingcounts
#sighting counts dataframe to count the frequency of sightings in each state.
```

```python
population=UFO.drop_duplicates('state')
#population=UFO[['state',"POPESTIMATE2019"]].unique()
population[['state',"POPESTIMATE2019"]]
```

```python
sightingvspop=sightingcounts.merge(population,on='state')
sightingvspop
#A new table was created to compare sightings vs population in a scatterplot for the next cell.
```

```python
sns.scatterplot(data=sightingvspop,x='count',y="POPESTIMATE2019")
#population in millions on y axis,count on x axis
#As the population goes up,so do sightings.
```
