## Bar Graph using matplotlib

### Use Pandas `pivot_table` function (https://pbpython.com/pandas-pivot-table-explained.html) to find the total number of orders for each category of the food item:

```python
table = pd.pivot_table(data = df, index = 'category', 
                       values = 'num_orders', aggfunc = np.sum)
```

```python
help(pd.pivot_table)
```

```python
table
```

### A bar graph can be generated using plt.bar() in matplotlib:

```python
help(table.index)
```

```python
help(plt.bar)
```

```python
# bar graph itself
plt.bar(table.index,table['num_orders'])

# xticks 
plt.xticks(rotation = 75)

# x and y-axis labels
plt.xlabel('Food item')
plt.ylabel('Quantity sold')

plt.savefig('foodplot4', dpi= 200)
plt.show()
```

```python
# dictionary for cuisine and their total orders
d_cuisine = {}

# total number of orders
total = df['num_orders'].sum() # total number of orders 
for i in range(df['cuisine'].nunique()): # iterate through each unique
    
    #cuisine's label 
    c = df['cuisine'].unique()[i]
    #print(c)
    
    # number of orders for the cuisine
    c_order = df[df['cuisine']==c]['num_orders'].sum()  # summing 
    d_cuisine[c] = c_order/total # get the percentage 
```

```python
d_cuisine
```

```python
# makes pie plot -- uses the counts but makes them into %s 
plt.pie([x*100 for x in d_cuisine.values()],
       labels = [x for x in d_cuisine.keys()])

# label
plt.title("Cuisine share")
plt.savefig('pieplot')

plt.show()
```

### Displaying values on your chart with `autopct` parameter 
>- autopct enables you to display the percent value using Python string formatting. For example, if autopct='%. 2f' , then for each pie wedge, the format string is '%. 2f' and the numerical percent value for that wedge is pct , so the wedge label is set to the string '%

### Making pieces stand out on your chart with the `explode` parameter
>- the explode parameter can be used to offset a wedge to make it stand out from the rest (for example, Italian can be offset by passing in an offset value)

```python

# makes pie plot -- uses the counts but makes them into %s 
plt.pie([x*100 for x in d_cuisine.values()],
       labels = [x for x in d_cuisine.keys()],
       autopct = '%0.1f', explode = [0, 0, 0.1, 0])

# label
plt.title("Cuisine share")
plt.savefig('pieplot')

plt.show()
```

# Visualize

## The scatterplot() is Seaborn's default relational plot you can create with `relplot()` 
>- Can also be forced by setting kind="scatter"

```python
sns.set(style="darkgrid")
sns.relplot(x="total_bill", y = "tip", data = tips)
```

```python
help(sns.relplot)
```

You can use the `hue` parameter to determine which column in the data frame should be used for colour encoding. ... 
>- Adding `hue="smoker"` tells seaborn you want to colour the data points for smoker and non-smoker differently.

```python
sns.relplot(x="total_bill", y = "tip", 
            hue = "smoker", data = tips)
```

## Create BoxPlots of all numerical columns in the `tips` dataset
>- General Syntax: `sns.boxplot(data = DataFrame, options)`
>- Common options
>>- `showmeans=True` will display a mean symbol
>>- `palette = <seabornpalettes>`, [seaborn color palettes](https://seaborn.pydata.org/tutorial/color_palettes.html)
>>- [Seborn Color Reference](https://www.codecademy.com/articles/seaborn-design-ii)


```python
sns.boxplot(tips['total_bill'],
           showmeans=True, palette = 'pastel')
```

```python
sns.boxplot(tips['tip'], showmeans=True, palette = 'colorblind')
```

```python
sns.boxplot(tips['size'], showmeans=True, palette='bright')
```

## Create a Distribution Plot
>- Use `distplot()` for the total_bill data
>- General Syntax: `sns.distplot(column, options)`
>- Optional: `bins=<integer>` allows you to specify how wide your bars are

```python
sns.distplot(tips['total_bill'])
```

## Use `jointplot()` to compare two variables
>- Question: How does the `total_bill` relate to `tip`?
>- General syntax: `jointplot(x = 'column1', y='column2', data=DataFrame)`
>- Optional: you can add `kind='reg'` to show a regression line on your plot

```python
sns.jointplot(x=tips['total_bill'],
             y=tips['tip'], kind = 'reg')
```
