## Problem 2
Given a weight in grams gramWeight, calculate the number of kilograms and grams. Your function should print the following:

> *gramWeight* is *X* kilograms and *Y* grams

where gramWeight is the number of grams passed to the function, X is the number of kilograms in gramWeight, and Y is the leftover grams. 

#### Write your code under the following function header: `def calcGramsAndKg(gramWeight):`

````python
def calcGramsAndKg(gramWeight): # this is the function header; function takes one parameter 
    if (gramWeight<0):
        print("Invalid number of grams")
    else:
        X=gramWeight//1000
        Y=gramWeight%1000
        print(gramWeight," is ",X,"Kilograms and ",Y,"Grams")

#To calculate grams to kilograms the gramweight was divided by 1000.
#Modulo was used for the quotient to get the code to match the format _ kilograms and _ grams on the next line.
   
````
