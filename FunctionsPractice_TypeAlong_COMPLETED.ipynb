# Creating a `Magic 8 Ball` Program With a Function
Task: Create a Magic 8 Ball program that randomly returns the various answers of the Magic 8 Ball game to the screen. 
1. Create a function called `getAnswer` that has the parameter variable, `answerNumber`
2. Pick 9 of the possible responses from the Magic 8 Ball designed by Mattel in the 1950s
3. Design a control flow in your `getAnswer` function that will output the various Magic 8 Ball text
4. Create a variable, `randNum`, that is assigned random numbers from 1-9
5. Print a user's fortune when the function is called

```python
import random 

def getAnswer():
    question = input("Ask the Magic 8 ball a question: ")
    randNum = random.randint(1,9) # gives a random integer in the range that you specify 
    if (question == "q"):
        quit()
    elif randNum == 1:
        print("It is certain")
    elif randNum == 2:
        print("Reply hazy, try again")
    elif randNum == 3:
        print("Yes - definitely")
    elif randNum == 4:
        print("Cannot predict now")
    elif randNum == 5:
        print("Don't count on it")
    elif randNum == 6:
        print("Most likely")
    elif randNum == 7:
        print("Outlook not so good")
    elif randNum == 8:
        print("Signs point to yes")
    elif randNum == 9:
        print("Very doubtful")
```

```python
getAnswer()
```

### Write a function for deciding if you should wear a mask. 

```python
def wearMask(goingOut, isSick): # defining a function that takes 2 parameters - going out and sick
    if goingOut == True: # if you're going out 
        print("Wear your mask!!!")
    else: # if you're not going out
        if isSick is True: # and you're sick
            print("wear your mask") # wear your mask still
        else: # if you aren't sick and not going out
            print("you don't have to wear a mask at home")
```

```python
wearMask(True, False)
wearMask(False, True)
wearMask(False, False)
```

### Word Capitalization: Create a function, that when passed a word checks if the word starts with an uppercase letter, prints "Already capitalized" if it does, and changes it and prints the capitalized version if it doesn't.  

```python
# index is the spot of an item 
def cap(word):
    new = ""
    if word[0].isupper(): # checks if the first letter of the word is upper case
        print("Already capitalized")
    else: 
        new = word.upper() # .capitalize() capitalizes string 
        print(new)
```

```python
cap('hello')
```

### Your *own* Absolute Value function: The built-in abs() function in Python returns the absolute value of a given number. Write a function that prints the absolute value of any number passed to it. 

```python
def absolute_value(num):
    if num >= 0: # if the number is positive 
        return num # the absolute value is that number 
    else:
        return -num
```

```python
print(absolute_value(-4))
```
