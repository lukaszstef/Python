## Mapping Functions 
## Using map(), apply(), applymap()
Source: https://medium.com/aikiss/pandas-apply-and-map-53868c3a0699

>- DataFrame.apply operates on entire rows or columns at a time.

>- DataFrame.applymap, Series.apply, and Series.map operate on one element at time.

>- Series.apply and Series.map are similar and often interchangeable. 

## Using `apply()` to apply a function along an axis of the DataFrame or on values of a series

### Write a function that multiplies all numbers in column 'Sales' by 1000 and stores the results in a new column named, `'Actual_Sales'`

```python
#use a lambda function to apply a function to every row in the data frame
future['Actual_Sales']=future.apply(lambda row: row.Sales *1000, axis=1)
```

### Axis parameter is the axis along which the function is applied:

>- 0 or ‘index’: apply function to each column.

>- 1 or ‘columns’: apply function to each row.
### axis = {0 or *index*, 1 or *columns*}, *default* 0


```python
future
```

### Then, use `apply()` to apply your function across the columns
>- `axis = 1` tells apply() to work across the columns

```python
lowerizer=lambda x: x.lower()
future['Restaurant'].apply(lowerizer)
```

# We use `applymap()` to apply functions across every element in a DataFrame
