### Practice Problem 1: Calculate if solar energy can power average house

Solar energy, which is measured in kilowatt-hours (kWh), is generated by a solar panel or array of panels. The amount of energy generated by a solar panel depends on its area, efficiency, solar radiation of the location, and performance ratio of the array. The equation used to calculate solar energy output is: 

E = A * r * H * PR

**E is the average energy generated annually by the solar array in kWh.** The parameters on the right side in this equation are: 

>- A: the area of the solar array 
>- r: the solar array efficiency 
>- H: the average solar radiation 
>- PR: the panel’s performance ratio, which is usually between 0.5 and 0.9, with a default of 0.75

Use the following function header:

`def calcSolarOutput(A, r, H):`

```python
def calcSolarOutput(A, r, H): # function header -- takes three parameters 
    energy = A * r * H * .75 # this is the calculated energy - check if enough
    enough = "" # initialize an output string that will change based on energy output
    if (energy>= 8500): # if energy is at least 8500 kwh 
        enough = "is" # add the word "is" to our output statement
    else:
        enough = "is not"
    
    output = "The average annual electricity ourput for this array is EkWh. The array is "  + str(energy)+ " kWh. The array " + enough + " sufficient to supply enough electricity to an average home in Colorado." 
    print(output)
```

### Practice Problem 2: Snacks & drinks 
For each day, you will eat X bags of snack and 2X bottles of drink. Write a function that takes two string parameters snack and drink, and a number X, then print snack for X times in a single line and print drink for 2X times in another single line. 

For example, given snack="Chips", drink="Water", and X=5, your function should print:

ChipsChipsChipsChipsChips

WaterWaterWaterWaterWaterWaterWaterWaterWaterWater


```python
def snacksAndDrinks(snack, drink, X):
    snackString = ""
    drinkString = ""
    # creating empty strings for the snack and drink text 
    for i in range(X): # go through the length of X (so 5 times for example)
        snackString += snack # add snack text once to the snackString
        drinkString += drink + drink  # add drink text twice to the drinkString
    print(snackString)
    print(drinkString)
```

```python
snacksAndDrinks("chips", "water", 5)
```

```python
def snacksDrink(snack,drink,x):
    print(snack*x)
    print(drink*2*x)
```

```python
snacksDrink("chips", "water", 5)
```

### Practice Problem 3: Practice with range( )
Given two numbers startNum and endNum, and a Boolean doAverage, write a function that does the following: if doAverage is True, calculate and print the average of numbers from startNum to but not including endNum; If doAverage is False, print the numbers from startNum to and including endNum in single line with a space separating the numbers. 


For example, given startNum = 10, endNum = 13, and doAverage = True, your function should print
11.0

given startNum = 10, endNum = 13, and doAverage = False, your function should print
10 11 12 13 

```python
def numbersAverage(startNum, endNum, doAverage):
    if(doAverage):
        # calculate the average
        sum = 0 # sum variable to keep track of the sum of numbers
        # make a counter variable to keep track of total count of numbers
        count = 0  # this is our counter variable
        for i in range(startNum, endNum):
            sum += i 
            # adding the value of the iterator to the sum
            count += 1 # add one to the count for each iteration 
        print(sum/count)
    else: # do average is false 
        # create a blank string that we will add to
        string = ""
        for i in range(startNum, endNum+1):
            string += str(i) + " " # each pass through the loop, we add the iterator to the string variable 
        print(string) #print output 
```

```python
numbersAverage(10, 13, True)
```

### Practice Problem 4: Summation

Starting at 1 and increasing by 5, add up all the numbers between 1 and 100. Use a variable to store the and display the summation for each iteration through the loop. 


```python
totalSum = 0
for i in range(1,100,5):
    totalSum = totalSum + i
    print("i is" + str(i))
    print("The current total sum " + str(totalSum))
print("The total sum is " + str(totalSum))
    
```
