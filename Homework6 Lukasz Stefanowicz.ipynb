## 2. Pull out all of the different `Round` names and store them in a list (you can call the `.tolist()` function to convert the extracted object.) 

```python
#get unique rounds and set to list
round_names = jep[' Round'].unique().tolist()
round_names
```

## 3. Make a dictionary with the round names and their respective top categories.

```python
d = {}
# loop through unique rounds and add top category of each round to dict
for item in round_names:
    d[item] = jep.loc[jep[' Round'] == item][' Category'].mode()[0]
```

## 5. Do "cats" or "dogs" come up more frequently in the questions? 
* Hint: `.str.contains()` can be used on the questions you get from the data

```python
# store total of cats and dogs occurences, then compare the two
cats = jep[' Question'].str.contains("cats").sum()
dogs = jep[' Question'].str.contains("dogs").sum()
if cats > dogs:
    print("Cats")
else:
    print("Dogs")
```
