`def makeADictionary(listOfStrings):`

`listOfStrings` = [“coffee”,”bananas”,”pickled eggs”]

{“coffee”:6, ”bananas”:7, “pickled eggs”:12}

**(1)** Write a function that takes a string of mixed upper and lower characters and returns the string characters such that the lowercase letters come first.

```python
def stringmixed(string):
    finalstring=''
    for character in string:
        if character.islower():
            finalstring=character+finalstring
        else:
            finalstring+=character
    return finalstring


stringmixed('KQwesDCv')



```

**(2)** Write a function that counts all lower case, upper case, digits, and special symbols in a given string.

```python
def countsall(string):
    numlower=0
    numUpper=0
    numDigits=0
    numSymbols=0
    for c in string:
        if c.islower():
            numlower+=1
        elif c.isupper():
            numUpper+=1
        elif c.isdigit():
            numDigits+=1
        else:
            numSymbols+=1
    print(numlower)
    print(numUpper)
    print(numDigits)
    print(numSymbols)
countsall('LkdJ730#@%dk2Q')




```

```python
def fullName(string):
    names=string.split()
    print(names[0][0]+"."+names[1][0]+"."+names[2])

fullName('Neli Dianna Radpour')
```
